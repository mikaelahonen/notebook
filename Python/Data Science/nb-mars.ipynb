# Multivariate Adaptive Regression Spline
MARS method fits multiple linear models to data.

[Read the docs](https://contrib.scikit-learn.org/py-earth/content.html).

## Example in two dimensional space

```python
import numpy as np
import pandas as pd
from pyearth import Earth
from matplotlib import pyplot as plt

#Define number of data points
n3 = 30
n1 = n3/3

#Create data
x = np.arange(n3)
y_1 = np.arange(n1)
y_2 = np.repeat(n1, n1)
y_3 = np.arange(n1, n3-n1)
y = np.concatenate((y_1, y_2, y_3), axis=None)

#Fit an Earth model
model = Earth()
model.fit(x, y)

#Print the model
#print(model.trace())
#print(model.summary())

#Predict
y_pred = model.predict(x)

#Plot the model
plt.plot(x, y, 'r-', label='Original')
plt.plot(x, y_pred, 'b.', label='Predicted')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Original VS Predicted by MARS')
plt.legend()
plt.show()
```
