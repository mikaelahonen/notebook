# Apriori
Apriori is used to find frequent itemsets from data. It takes only two arguments: the data and the treshold about what is considered as frequent.

The algoritm first looks itemsets of size one. If the frequency in the records is higher than the treshold, the product will be involved in the next iteration.

Next, all itemsets of size two containing a product that exceeded the treshold in previous stage will be under investigation.

The process continues until no frequent itemsets are found or maximum combination length has been reached.

[Wikipedia has an example](https://en.wikipedia.org/wiki/Apriori_algorithm) that is easy to understand. The example for this code taken from `mlextend` tutorial [here](http://rasbt.github.io/mlxtend/user_guide/frequent_patterns/apriori/#api).

## Import

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
```

## Pre process data

```python
data_values = [
    ['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
    ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
    ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
    ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
    ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']
]

#Turn the values to binaries
te = TransactionEncoder()
data_binary = te.fit(data_values).transform(data_values)
#Turn binary data to data frame
df = pd.DataFrame(data_binary, columns=te.columns_)
df.head(10)
```

## Run apriori

```python
#Get apriori results in data frame
df_apr = apriori(df, min_support=0.6, use_colnames=True)
df_apr
```
