# PCA
Principal component analysis. PCA transforms data so that new features will be created.

First use `sklearn`'s `decomposition.PCA()` to create an pca object.

Fit the PCA transform with training data using pca object's `fit()` function and later transform the test data with `transofrm()` function.

```python
from sklearn import linear_model, decomposition, datasets
import numpy as np
import matplotlib.pyplot as plt
```

## Create initial data and plot

```python
x_max = 10

#Values from 0 to 10 every 0.2
x1 = np.arange(0, x_max, .2)
#Make x2 somewhat linear with x1 by adding random noise to x1
x2 = x1 + np.random.normal(loc=0, scale=1, size=len(x1))
#Get some imaginary colors
color = np.where((x1>x_max/2) | (x2>x_max/2), "blue", "red")

plt.scatter(x1, x2, color=color,marker="o",alpha=0.5)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

## Do PCA tranformation 

```python
#Turn arrays to a row wise list. og=original
rows_og = np.c_[(x1, x2)]
#Print first data points
print("\nOriginal data in sklearn compatible format")
print(rows_og[0:5])

#Creat a PCA object
pca = decomposition.PCA(n_components=2)
#Do thePCA  transformation. tf=transform
rows_tf = pca.fit_transform(rows_og)

#Show first data of transformed data
print("\nTransformed data")
print(rows_tf[0:5])

#Show how many percentages each variables explains variance
print("\nVariance explained")
print(pca.explained_variance_ratio_)
```

## Plot transformed data

```python
x1_tf = []
x2_tf = []
x_zero = []
for r in rows_tf:
    x1_tf.append(r[0])
    x2_tf.append(r[1])
    x_zero.append(0)


#Plot both PCA components
plt.scatter(x1_tf, x2_tf, color=color,marker="o",alpha=0.5)
plt.title("Both PCA components")
plt.xlabel("var 1")
plt.ylabel("var 2")
plt.show()

#Plot only the most significant component
plt.scatter(x1_tf, x_zero, color=color,marker="o",alpha=0.5)
plt.title("Only the most significant PCA component")
plt.xlabel("var 1")
plt.show()
```
