```python
import matplotlib.pyplot as plt

x = range(10)
y1 = [12,4,16,43,32,32,22,5,7,24]
y2 = [9,3,12,40,20,30,8,5,6,20]
```

## Two area charts

```python
fig, ax = plt.subplots(figsize=(10,5))
#Plot 1
ax.fill_between(x, y1, 0, color="blue", alpha=0.5)
#Plot 2
ax.fill_between(x, y2, 0, color="yellow", alpha=0.5)
#Show
plt.show()
```

## Combined bar and line chart with individual axes

```python
fig, ax1 = plt.subplots()

#Plot 1
ax1.bar(x, y1, alpha=0.5)
ax1.set_xlabel("X lab")
ax1.set_ylabel('y1')

#Plot 2
ax2 = ax1.twinx()
ax2.plot(x, y2, color="green")
ax2.set_ylabel('y2')

plt.show()
```

```python
import numpy as np
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# Note that using plt.subplots below is equivalent to using
# fig = plt.figure and then ax = fig.add_subplot(111)
fig, ax = plt.subplots()
ax.plot(t, s, color="green")

ax.set(xlabel='Time', ylabel='Y axis', title='Some Curve')
ax.grid()

plt.show()
```

## Annotation

```python
plt.plot(x, y1)
plt.annotate('Drop starts', xy=(5, 32), xytext=(6, 40),arrowprops=dict(facecolor='black', shrink=0.05),)
plt.show()
```

## Rectangle
Rectanlge((x_min, y_min), x_width, y_height)

```python
from matplotlib.patches import Rectangle

plt.plot(x, y1)
rect_1 = Rectangle((1, 5), 2, 40, alpha=0.3, color='green')
rect_2 = Rectangle((3, 5), 4, 40, alpha=0.3, color='red')
plt.gca().add_patch(rect_1)
plt.gca().add_patch(rect_2)
plt.show()
```
