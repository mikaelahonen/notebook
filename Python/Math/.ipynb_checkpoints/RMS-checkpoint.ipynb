## RMS
RMS = Root Mean Square.

It is used instead of arithmetic mean in some cases.
RMS suits well to express mean of data that is symmetric around X-axis.
In these situations regular arithmetic mean would be zero.
Vibration data is a good example.


```python
import matplotlib.pyplot as plt
import matplotlib.lines as lines
import numpy as np

# Data
y = [0,4,6,7,6,4,0,-4,-6,-7,-6,-4,0]
y_len = len(y)-1
```

```python
#Plot data
fig, ax = plt.subplots()
ax.plot([0,y_len], [0,0], color='gray', linestyle='-', linewidth=1)
ax.plot(y, color='k')
plt.show()
```

```python
#Arithemtic mean
mean_arithmetic = np.mean(y)
print("Arithmetic mean")
print(mean_arithmetic)

#Arithemtic mean with absolute values
mean_arithmetic_abs = np.mean(np.absolute(y))
print("Arithmetic mean with absolute values")
print(mean_arithmetic_abs)

#RMS
mean_rms = np.sqrt(np.mean(np.square(y)))
print("RMS")
print(mean_rms)
```

```python
#Plot data with different means
fig, ax = plt.subplots()
ax.plot([0, y_len], [mean_arithmetic, mean_arithmetic], color='r', linestyle='-', linewidth=2, label="Arithmetic")
ax.plot([0,y_len], [mean_arithmetic_abs, mean_arithmetic_abs], color='g', linestyle='-', linewidth=2, label="Arithmetic - Abs")
ax.plot([0,y_len], [mean_rms, mean_rms], color='b', linestyle='-', linewidth=2, label="RMS")
ax.plot(y, color='k')
plt.legend()
plt.show()
```
